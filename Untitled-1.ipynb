```python
import cv2
import os
from ultralytics import YOLO

# Load your trained YOLO model
model = YOLO(r'C:\Users\findm\OneDrive\Desktop\model1\runs\detect\helmet_detection_new\weights\best.pt')

# Output directory
output_dir = r'C:\Users\findm\OneDrive\Desktop\model1\results'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Process a video and save the annotated video output
def process_video(video_path):
    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        print("Error: Could not open video.")
        return

    # Get video properties
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for .mp4 output

    # Output video path
    video_name = os.path.splitext(os.path.basename(video_path))[0]
    output_video_path = os.path.join(output_dir, f"{video_name}_output.mp4")

    # Define VideoWriter object
    out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

    frame_idx = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        results = model(frame)
        result_frame = results[0].plot()  # Draw bounding boxes on the frame

        out.write(result_frame)  # Write annotated frame to output video
        frame_idx += 1
        print(f"Processed frame {frame_idx}")

    cap.release()
    out.release()
    print(f"Video processing complete. Output saved to {output_video_path}")



video_path = r'126434-735976920_small.mp4'  # Update this with your actual video path
process_video(video_path)
```

```python
import os
import cv2
from ultralytics import YOLO

# Load the YOLO model (adjust the path accordingly)
model = YOLO(r"C:\Users\findm\OneDrive\Desktop\model1\runs\detect\helmet_detection_new\weights\best.pt")

# Function to extract and save the number plate only if there is no helmet and a number plate inside the rider's bounding box
def extract_and_save_number_plate(image_path, output_folder, confidence_threshold=0.5):
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error: Unable to read image from {image_path}")
        return

    # Predict with YOLO
    results = model.predict(source=image, conf=confidence_threshold)

    # Lists to hold detections
    rider_boxes = []
    helmet_boxes = []
    no_helmet_boxes = []
    plate_boxes = []

    for result in results:
        for box in result.boxes:
            cls = int(box.cls[0])
            conf = float(box.conf[0])
            x1, y1, x2, y2 = map(int, box.xyxy[0])

            if conf >= confidence_threshold:
                if cls == 3:
                    rider_boxes.append((x1, y1, x2, y2))
                elif cls == 2:
                    helmet_boxes.append((x1, y1, x2, y2))
                elif cls == 1:
                    no_helmet_boxes.append((x1, y1, x2, y2))
                elif cls == 0:
                    plate_boxes.append((x1, y1, x2, y2))

    # Now check each rider box
    for rx1, ry1, rx2, ry2 in rider_boxes:
        has_helmet = any(hx1 >= rx1 and hy1 >= ry1 and hx2 <= rx2 and hy2 <= ry2 for hx1, hy1, hx2, hy2 in helmet_boxes)
        has_no_helmet = any(nx1 >= rx1 and ny1 >= ry1 and nx2 <= rx2 and ny2 <= ry2 for nx1, ny1, nx2, ny2 in no_helmet_boxes)
        matching_plate = None

        for px1, py1, px2, py2 in plate_boxes:
            if px1 >= rx1 and py1 >= ry1 and px2 <= rx2 and py2 <= ry2:
                matching_plate = (px1, py1, px2, py2)
                break  # Take only one plate per rider

        # Save only if the rider has no helmet AND a plate inside their box AND does NOT have a helmet
        if has_no_helmet and matching_plate and not has_helmet:
            px1, py1, px2, py2 = matching_plate
            cropped_plate = image[py1:py2, px1:px2]
            plate_filename = f"plate_nohelmet_{rx1}_{ry1}.jpg"
            save_path = os.path.join(output_folder, plate_filename)
            cv2.imwrite(save_path, cropped_plate)
            print(f"Saved Number Plate: {save_path}")

# Main execution
if __name__ == "__main__":
    input_image = r"C:\Users\findm\OneDrive\Desktop\model1\new118.jpg"  # Replace with your test image path
    output_folder = r"C:\Users\findm\OneDrive\Desktop\model1\results"

    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    extract_and_save_number_plate(input_image, output_folder, confidence_threshold=0.4)
```

```python
import cv2
from paddleocr import PaddleOCR
import re
import os
from datetime import datetime

# ---------- Step 1: Resize the image ----------
def resize_image(input_path, output_path, width=500, height=200):
    img = cv2.imread(input_path)
    resized = cv2.resize(img, (width, height))
    cv2.imwrite(output_path, resized)
    return output_path

# ---------- Step 2: OCR and Save Result ----------
def extract_number_plate(image_path, save_file_path):
    # Resize image
    resized_path = resize_image(image_path, "resized_plate.jpg")

    # Initialize OCR
    ocr = PaddleOCR(use_angle_cls=True, lang='en')

    # Run OCR
    results = ocr.ocr(resized_path, cls=True)

    # Process results
    if results and results[0]:
        detected_texts = []
        for line in results[0]:
            text = line[1][0]
            cleaned_text = text.strip().replace(" ", "").upper()
            detected_texts.append(cleaned_text)

        print("🔍 Raw OCR Output:", detected_texts)

        # Combine all text parts into one string
        combined_text = ''.join(detected_texts)

        # Apply regex to extract plate-like format
        match = re.search(r"[A-Z]{2}[0-9]{2}[A-Z]{1,3}[0-9]{4}", combined_text)
        number_plate = match.group() if match else "NOT FOUND"

        

        # Save to file
        with open(save_file_path, "a") as f:
            now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            image_name = os.path.basename(image_path)
            f.write(f"{now} - {image_name} - {number_plate}\n")

        print(f"📁 Saved to: {save_file_path}")
        return number_plate

    else:
        print("⚠️ No text detected in the image.")
        return None

# ---------- Step 3: Set your file paths ----------
input_path = r"C:\Users\findm\OneDrive\Desktop\model1\results\plate_nohelmet_7_21.jpg"
save_file = r"C:\Users\findm\OneDrive\Desktop\model1\results\detected_numbers.txt"

extract_number_plate(input_path, save_file)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def preprocess_plate(image_path):
    # Load the image
    img = cv2.imread(image_path)

    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Apply Gaussian Blur to reduce noise
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)

    # Apply OTSU thresholding (better than adaptive)
    _, binary = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Invert colors (make text black, background white)
    binary = cv2.bitwise_not(binary)

    # Display result
    plt.imshow(binary, cmap='gray')
    plt.title("Thresholded Image")
    plt.show()

    return binary

# Run the preprocessing
processed_plate = preprocess_plate(r"C:\Users\findm\OneDrive\Desktop\model1\results\plate_60_44.jpg")
```

```python
import cv2
import matplotlib.pyplot as plt

def preprocess_image(image_path):
    # Load image
    img = cv2.imread(image_path)
    img = cv2.resize(img, (400, 150))  # Resize for consistency

    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Apply bilateral filter to reduce noise
    filtered = cv2.bilateralFilter(gray, 11, 17, 17)

    # Threshold to get binary image
    _, thresh = cv2.threshold(filtered, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

    # Show result
    plt.imshow(thresh, cmap='gray')
    plt.title("Preprocessed Image")
    plt.axis('off')
    plt.show()

    return thresh, img

# Example usage
image_path = r"C:\Users\findm\OneDrive\Desktop\model1\results\plate_359_104.jpg"
thresh_img, original_img = preprocess_image(image_path)
```
